```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Scrapping structure

Product_Name = []
Product_Links = []
Product_Rating = []
Product_Reviews = []
Product_Price = []

for j in range(1, 20):
    URL = "https://www.amazon.in/s?k=smartphone&crid=3GK5FSN2G3LR9&qid=1726687525&sprefix=%2Caps%2C228&ref=sr_pg_{}".format(j, j)
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

    Outer_Page = requests.get(URL, headers=HEADERS)
    Soup = BeautifulSoup(Outer_Page.content, 'html.parser')

    Each_Div = Soup.find_all('div', attrs = {'class': 'puis-card-container s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v9jpl3r55yfpi2nize9atlo87q s-latency-cf-section puis-card-border'})



    for contents in Each_Div:
        try:
            Extr_Names = contents.find('span', attrs = {'class': 'a-size-medium a-color-base a-text-normal'}).text.strip()
        except(AttributeError, IndexError):
            Extr_Names = " "
        Product_Name.append(Extr_Names)
    
        try:
            Extr_Links = "https://www.amazon.in" + contents.find('a', attrs = {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'}).get('href')
        except(AttributeError, IndexError):
            Extr_Links = " "
        Product_Links.append(Extr_Links)
    
        try:
            Extr_Rating = contents.find('span', attrs = {'class': 'a-icon-alt'}).text.strip()
        except(AttributeError, IndexError):
            Extr_Rating = " "
        Product_Rating.append(Extr_Rating)

        try:
            Extr_Reviews = contents.find('span', attrs = {'class': 'a-size-base s-underline-text'}).text.strip()
        except(AttributeError, IndexError):
            Extr_Reviews = " "
        Product_Reviews.append(Extr_Reviews)

        try:
            Extr_Prices = contents.find('span', attrs = {'class': 'a-price-whole'}).text.strip()
        except(AttributeError, IndexError):
            Extr_Prices = " "
        Product_Price.append(Extr_Prices)    
```

### Creating DataFrame

```python
Data = {'Product_Name': Product_Name,
        'Link': Product_Links,
        'Rating': Product_Rating,
        'Reviews': Product_Reviews,
        'Price': Product_Price,
        }

Amazon_Data = pd.DataFrame(Data)
Amazon_Data
```

```python
Amazon_Data.replace(" ", np.nan, inplace = True)
```
